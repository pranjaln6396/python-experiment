```python
# 1. Count and display number of capital letters in a string.
string = input("Enter string: ")
count = 0
for ch in string:
    if ch.isupper():
        count += 1
print("Capital letters:", count)
```

```python
# 2. Count total number of vowels in a string.
string = input("Enter string: ")
vowels = "aeiouAEIOU"
count = 0
for ch in string:
    if ch in vowels:
        count += 1
print("Total vowels:", count)
```

```python
# 3. Input a sentence and print words in separate lines.
sentence = input("Enter sentence: ")
words = sentence.split()
for word in words:
    print(word)
```

```python
# 5. Count occurrence of each alphabet (case insensitive).
string = input("Enter string: ").upper()
checked = set()

for ch in string:
    if ch.isalpha() and ch not in checked:
        print(string.count(ch), ch)
        checked.add(ch)
```

```python
# 6. Count number of unique words using sets.
sentence = input("Enter sentence: ")
words = sentence.split()
unique_words = set(words)
print("Unique words count:", len(unique_words))
```

```python
# 7. Create 2 sets of fruits and perform operations.
n = int(input("Enter number of fruits in each set: "))

s1 = set()
s2 = set()

print("Enter fruits for set 1:")
for i in range(n):
    s1.add(input())

print("Enter fruits for set 2:")
for i in range(n):
    s2.add(input())

print("a) Common fruits:", s1 & s2)
print("b) Only in s1:", s1 - s2)
print("c) Total fruits count:", len(s1 | s2))
```

```python
# 8. Set operations on given sets.
S1 = {"Red", "yellow", "orange", "blue"}
S2 = {"violet", "blue", "purple"}

print("Union:", S1 | S2)
print("Intersection:", S1 & S2)
print("Difference S1-S2:", S1 - S2)
print("Difference S2-S1:", S2 - S1)
print("Symmetric Difference:", S1 ^ S2)
```
