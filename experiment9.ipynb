```python
#create a class student with attibute name,name,sapid,makrs(phy,chem,maths) create 3 boject using user input and display their details

class Student:
    def input_data(self):
        self.name = input("Enter name: ")
        self.sap_id = input("Enter SAP ID: ")
        self.phy = int(input("Enter Physics marks: "))
        self.chem = int(input("Enter Chemistry marks: "))
        self.math = int(input("Enter Maths marks: "))

    def display(self):
        print("Name:", self.name)
        print("SAP ID:", self.sap_id)
        print("Physics:", self.phy)
        print("Chemistry:", self.chem)
        print("Maths:", self.math)
        print()

students = []

for i in range(3):
    s = Student()
    s.input_data()
    students.append(s)

for s in students:
    s.display()
```

```python
#Add a constructor and implement methods:display(),marks_percentage(),result(),Function to find class average

class Student:

    def __init__(self,name,sap,phy,chem,math):
        self.name = name
        self.sap = sap
        self.phy = phy
        self.chem = chem
        self.math = math

    def display(self):
        print("Name:",self.name)
        print("SAP:",self.sap)
        print("Marks:",self.phy,self.chem,self.math)

    def marks_percentage(self):
        per = (self.phy+self.chem+self.math)/3
        return per

    def result(self):
        if self.phy>40 and self.chem>40 and self.math>40:
            print("Result: Pass")
        else:
            print("Result: Fail")

students = []

n = int(input("Enter number of students: "))

for i in range(n):
    name = input("Enter name: ")
    sap = input("Enter SAP: ")
    phy = int(input("Physics: "))
    chem = int(input("Chemistry: "))
    math = int(input("Maths: "))

    s = Student(name,sap,phy,chem,math)
    students.append(s)

total = 0

for s in students:
    s.display()
    per = s.marks_percentage()
    print("Percentage:",per)
    s.result()
    total += per
    print()

print("Class Average:", total/n)
```

```python
#program to impliment different type of inheritance 
#single inheritance

class Parent:
    def show(self):
        print("This is parent class")

class Child(Parent):
    def display(self):
        print("This is child class")

obj = Child()
obj.show()
obj.display()
```

```python
#multiple iheritance 
class A:
    def showA(self):
        print("Class A")

class B:
    def showB(self):
        print("Class B")

class C(A,B):
    def showC(self):
        print("Class C")

obj = C()
obj.showA()
obj.showB()
obj.showC()
```

```python
#Create a class to demonstrate method overriding.

class Parent:
    def show(self):
        print("Parent method")

class Child(Parent):
    def show(self):
        print("Child method")

obj = Child()
obj.show()
```

```python
#Add two point objects.

class Point:

    def __init__(self,x,y):
        self.x = x
        self.y = y

    def __add__(self,other):
        x = self.x + other.x
        y = self.y + other.y
        return Point(x,y)

    def display(self):
        print("(",self.x,",",self.y,")")

p1 = Point(10,20)
p2 = Point(12,15)

p3 = p1 + p2

p3.display()
```
