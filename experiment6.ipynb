```python
# 1. Scan n values in range 0â€“3 and print occurrence of each value

n = int(input("Enter number of values: "))
values = []

for i in range(n):
    num = int(input("Enter value (0-3): "))
    if 0 <= num <= 3:
        values.append(num)
    else:
        print("Invalid input! Enter value between 0 and 3.")

# Count occurrences
for i in range(4):
    print(f"{i} occurred {values.count(i)} times")
```

```python
# 2. Create a tuple to store n numeric values and find average

n = int(input("Enter number of values: "))
numbers = []

for i in range(n):
    num = float(input("Enter number: "))
    numbers.append(num)

numbers_tuple = tuple(numbers)

average = sum(numbers_tuple) / len(numbers_tuple)

print("Tuple:", numbers_tuple)
print("Average:", average)
```

```python
# 4. Dictionary of persons

n = int(input("Enter number of persons: "))
students = {}

for i in range(n):
    name = input("Enter name: ")
    city = input("Enter city: ")
    students[name] = city

# a. Display all names
print("\nNames:")
for name in students.keys():
    print(name)

# b. Display all city names
print("\nCities:")
for city in students.values():
    print(city)

# c. Display student names and cities
print("\nStudent Details:")
for name, city in students.items():
    print(name, "lives in", city)

# d. Count number of students in each city
city_count = {}
for city in students.values():
    city_count[city] = city_count.get(city, 0) + 1

print("\nNumber of students in each city:")
for city, count in city_count.items():
    print(city, ":", count)
```

```python
# 5. Store movie details

n = int(input("Enter number of movies: "))
movies = {}

for i in range(n):
    print("\nEnter details for movie", i+1)
    name = input("Movie name: ")
    year = int(input("Year of release: "))
    director = input("Director name: ")
    cost = float(input("Production cost: "))
    collection = float(input("Collection made: "))

    movies[name] = {
        "year": year,
        "director": director,
        "cost": cost,
        "collection": collection
    }

# a. Print all movie details
print("\nAll Movie Details:")
for name, details in movies.items():
    print("\nName:", name)
    for key, value in details.items():
        print(key, ":", value)

#
```
