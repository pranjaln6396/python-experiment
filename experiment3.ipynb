```python
# 1. Check whether a given number is divisible by 4 and 9 both.
num = int(input("Enter number: "))
if num % 4 == 0 and num % 9 == 0:
    print("Divisible by both 4 and 9")
else:
    print("Not divisible by both 4 and 9")
```

```python
# 2. Check whether a given number is multiple of five or not.
num = int(input("Enter number: "))
if num % 5 == 0:
    print("Multiple of 5")
else:
    print("Not a multiple of 5")
```

```python
# 3. Find the greatest among two numbers.
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

if a > b:
    print("Greatest:", a)
elif b > a:
    print("Greatest:", b)
else:
    print("Numbers are equal")
```

```python
# 4. Find the greatest among three numbers (no two values same).
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a > b and a > c:
    print("Greatest:", a)
elif b > a and b > c:
    print("Greatest:", b)
else:
    print("Greatest:", c)

```

```python
# 5. Quadratic equation roots
a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))

d = b**2 - 4*a*c

if d > 0:
    print("Real and Distinct roots")
    r1 = (-b + d**0.5) / (2*a)
    r2 = (-b - d**0.5) / (2*a)
    print("Root1:", r1)
    print("Root2:", r2)
elif d == 0:
    print("Real and Equal roots")
    r = -b / (2*a)
    print("Root:", r)
else:
    print("Imaginary roots")
```

```python
# 6. Check leap year
year = int(input("Enter year: "))

if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print("Leap Year")
else:
    print("Not a Leap Year")

```

```python
# 7. Display next date
day = int(input("Enter day: "))
month = int(input("Enter month: "))
year = int(input("Enter year: "))

# Days in months
days_in_month = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]

# Check leap year for February
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    days_in_month[1] = 29

day += 1

if day > days_in_month[month - 1]:
    day = 1
    month += 1
    if month > 12:
        month = 1
        year += 1

print("Next Date:")
print("day =", day, "month =", month, "year =", year)

```

```python
# 8. Grade Sheet Program
name = input("Enter Name: ")
roll = input("Enter Roll Number: ")
sapid = input("Enter SAP ID: ")
sem = input("Enter Semester: ")
course = input("Enter Course: ")

print("Enter marks of five subjects:")
m1 = float(input("PDS: "))
m2 = float(input("Python: "))
m3 = float(input("Chemistry: "))
m4 = float(input("English: "))
m5 = float(input("Physics: "))

percentage = (m1 + m2 + m3 + m4 + m5) / 5
cgpa = percentage / 10

if 0 <= cgpa <= 3.4:
    grade = "F"
elif 3.5 <= cgpa <= 5.0:
    grade = "C+"
elif 5.1 <= cgpa <= 6.0:
    grade = "B"
elif 6.1 <= cgpa <= 7.0:
    grade = "B+"
elif 7.1 <= cgpa <= 8.0:
    grade = "A"
elif 8.1 <= cgpa <= 9.0:
    grade = "A+"
elif 9.1 <= cgpa <= 10:
    grade = "O"
else:
    grade = "Invalid"

print("\n----- Grade Sheet -----")
print("Name:", name)
print("Roll Number:", roll, "SAPID:", sapid)
print("Sem:", sem, "Course:", course)
print("Percentage:", percentage, "%")
print("CGPA:", cgpa)
print("Grade:", grade)
```
