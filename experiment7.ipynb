```python
# 1. Function to find maximum and minimum (without built-in functions)

def find_max_min(numbers):
    maximum = numbers[0]
    minimum = numbers[0]
    
    for num in numbers:
        if num > maximum:
            maximum = num
        if num < minimum:
            minimum = num
            
    return maximum, minimum


# Example
nums = [1000, 539, 2790, 84758, 364]
max_val, min_val = find_max_min(nums)
print("Maximum:", max_val)
print("Minimum:", min_val)
```

```python
# 2. Sum of cubes of numbers smaller than given number

def sum_of_cubes(n):
    total = 0
    for i in range(1, n):
        total += i ** 3
    return total


# Example
num = int(input("Enter a positive integer: "))
print("Sum of cubes:", sum_of_cubes(num))
```

```python


# 3. Print 1 to n using recursion

def print_numbers(n):
    if n == 0:
        return
    print_numbers(n - 1)
    print(n)


# Example
num = int(input("Enter value of n: "))
print_numbers(num)
```

```python
# 4. Fibonacci series using recursion

def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)


terms = int(input("Enter number of terms: "))
print("Fibonacci Series:")

for i in range(terms):
    print(fibonacci(i), end=" ")
```

```python
# 5. Lambda function for volume of cone

import math

volume_cone = lambda r, h: (1/3) * math.pi * r * r * h

radius = float(input("Enter radius: "))
height = float(input("Enter height: "))

print("Volume of Cone:", volume_cone(radius, height))
```

```python

# 6. Lambda function to return (max, min)

max_min = lambda lst: (max(lst), min(lst))

# Sample input
numbers = [10, 6, 8, 90, 12, 56]

result = max_min(numbers)
print("Output:", result)
```

```python
# a. Keyword Argument

def student(name, age):
    print("Name:", name)
    print("Age:", age)

# Calling using keyword arguments
student(age=20, name="ranjal")
```

```python
# b. Default Argument

def greet(name="Student"):
    print("Hello", name)

greet("pranjal")   # Provided argument
greet()          # Uses default value
```

```python
# c. Variable Length Argument

def add_numbers(*numbers):
    total = 0
    for num in numbers:
        total += num
    print("Sum:", total)

add_numbers(10, 20)
add_numbers(1, 2, 3, 4, 5)
```
