```python
# 8. Area of triangle when sides are given (Heron's formula)
a = float(input("Enter side 1: "))
b = float(input("Enter side 2: "))
c = float(input("Enter side 3: "))
s = (a + b + c) / 2
area = (s * (s - a) * (s - b) * (s - c)) ** 0.5
print("Area of triangle:", area)
```

```python
# 9. Convert seconds into hours, minutes and seconds
seconds = int(input("Enter seconds: "))
hours = seconds // 3600
minutes = (seconds % 3600) // 60
remaining_seconds = seconds % 60
print("Hours:", hours)
print("Minutes:", minutes)
print("Seconds:", remaining_seconds)
```

```python
# 10. Swap two numbers without additional variable
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
a, b = b, a
print("After swapping:")
print("a =", a)
print("b =", b)
```

```python
# 11. Sum of first n natural numbers
n = int(input("Enter value of n: "))
sum_n = n * (n + 1) // 2
print("Sum of first", n, "natural numbers:", sum_n)
```

```python
# 12. Truth table for bitwise operators (&, |, ^)
print("A B  A&B  A|B  A^B")
for A in [0, 1]:
    for B in [0, 1]:
        print(A, B, " ", A & B, "   ", A | B, "   ", A ^ B)
```

```python
# 13. Left shift and right shift
num = int(input("Enter number: "))
shift = int(input("Enter shift value: "))
print("Left Shift:", num << shift)
print("Right Shift:", num >> shift)
```
