```python
 #ques.1 Write a Python program to read names from name.txt and count the total number of names.

with open("name.txt", "r") as f:
    count = 0
    for line in f:
        if line.strip():      # ignore empty lines
            count += 1

print("Total names:", count)
```

```python
#Read names from a file and count how many names start with a vowel.

vowels = "AEIOUaeiou"
count = 0

with open("name.txt","r") as f:
    for name in f:
        name = name.strip()
        if name[0] in vowels:
            count += 1

print("Names starting with vowel:",count)
```

```python
#Write a program to find the longest name in a file.

with open("name.txt","r") as f:
    names = [line.strip() for line in f]

longest = names[0]

for name in names:
    if len(name) > len(longest):
        longest = name

print("Longest name:",longest)
```

```python
#Write a program to find the maximum number stored in a file.
try:
    with open("number.txt", "r") as f:
        nums = [int(line.strip()) for line in f]

    print("Maximum number:", max(nums))

except FileNotFoundError:
    print("File not found")
```

```python
#write a programe to calculate the average of the numbers in he file 

with open("number.txt","r") as f:
    nums = [int(line.strip()) for line in f]

avg = sum(nums)/len(nums)

print("Average:",avg)
```

```python
#wirte a programe to count the numbers greater then 100

with open("number.txt","r") as f:
    nums = [int(line.strip()) for line in f]

count = sum(1 for n in nums if n > 100)

print("Count of numbers greater than 100:", count)
```

```python
#write a prograe to perform integer division with exception handing 

try:
    a,b = input("Enter two numbers: ").split()

    a = int(a)
    b = int(b)

    print(a//b)

except ZeroDivisionError as e:
    print("Error Code:",e)

except ValueError as e:
    print("Error Code:",e)
```

```python
#write a programe to handle file not found exception

try:
    f = open("data.txt","r")
    print(f.read())

except FileNotFoundError:
    print("File not found")
```

```python
#create a custom exception for empty file 

class EmptyFileError(Exception):
    pass

try:
    with open("data.txt","r") as f:
        content = f.read()

        if content == "":
            raise EmptyFileError("File is empty")

        print(content)

except EmptyFileError as e:
    print(e)
```
