```python
# Q1) Find factorial of given number.
n = int(input("Enter number: "))
fact = 1
for i in range(1, n + 1):
    fact *= i
print("Factorial:", fact)
```

```python
# Q2) Check Armstrong number.
num = int(input("Enter number: "))
temp = num
power = len(str(num))
sum_val = 0

while temp > 0:
    digit = temp % 10
    sum_val += digit ** power
    temp //= 10

if sum_val == num:
    print("Armstrong Number")
else:
    print("Not Armstrong Number")
```

```python
# Q3) Fibonacci series up to given term.
n = int(input("Enter number of terms: "))
a, b = 0, 1
for i in range(n):
    print(a, end=" ")
    a, b = b, a + b
print()
```

```python
# Q4) Check prime number.
num = int(input("Enter number: "))
if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime Number")
else:
    print("Not Prime")
```

```python
# Q5) Check palindrome number.
num = input("Enter number: ")
if num == num[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
# Q6) Sum of digits.
num = int(input("Enter number: "))
sum_digits = 0
while num > 0:
    sum_digits += num % 10
    num //= 10
print("Sum of digits:", sum_digits)
```

```python
# Q7) Count and print numbers divisible by 5 or 7 between 1 to 100.
count = 0
for i in range(1, 200):
    if i % 5 == 0 or i % 7 == 0:
        print(i, end=" ")
        count += 1
print("\nTotal count:", count)
```

```python
# Q8) Convert lowercase to uppercase in a string.
string = input("Enter string: ")
print("Uppercase:", string.upper())
```

```python
# Q9) Print all prime numbers between 1 and 100.
for num in range(2, 101):
    for i in range(2, num):
        if num % i == 0:
            break
    else:
        print(num, end=" ")
print()
```

```python
# Q10) Print table for a given number.
num = int(input("Enter number for table: "))
for i in range(1, 11):
    print(num, "*", i, "=", num * i)
```
